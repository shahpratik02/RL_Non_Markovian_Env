```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_main=pd.read_csv(r'D:\RL PROJECT\rewards_main_algo_05-05.csv')
df_main
```

```python
plt.plot(df_main['scores'])
```

```python
df_only=pd.read_csv(r'D:\RL PROJECT\rewards_only_dqn_PO_lunar_07_05.csv')
df_only
```

```python
plt.plot(df_only['scores'])
```

```python
df_main['scores2']=[max(x,0) for x in df_main['scores']]
df_only['scores2']=[max(x,0) for x in df_only['scores']]
```

```python
df_main['moving_avg']=df_main.scores2.rolling(100,min_periods=1).mean()
df_only['moving_avg']=df_only.scores2.rolling(100,min_periods=1).mean()
```

```python
plt.rcParams['figure.figsize']=[10,8]

plt.plot(df_main['moving_avg'][:3000],label="Main Algorithm Moving Avg",color='blue')
plt.plot(df_only['moving_avg'][:3000],label="Only DQN Moving Avg",color='green')
plt.plot(df_only['scores'][:3000],label="Only DQN",linestyle='dashed',alpha=0.3,color='green')
plt.plot(df_main["scores"][:3000],label="Main ALgorithm",linestyle='dashed',alpha=0.3,color='blue')
plt.ylabel('rewards')
plt.xlabel('episodes')
plt.legend(loc="lower right")
```

```python
plt.plot(df_main['moving_avg'][:3000],label="Main Algorithm Moving Avg",color='blue')
plt.plot(df_only['moving_avg'][:3000],label="Only DQN Moving Avg",color='green')
plt.ylabel('rewards')
plt.xlabel('episodes')
plt.legend(loc="lower right")
plt.title("Main Algorithm vs DQN (Lunar Lander)")
plt.savefig("lunar_lander_new.png")
plt.show()
```
